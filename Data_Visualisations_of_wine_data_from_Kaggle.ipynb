## 1. Importing the required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

## 2. Loading the data into the data frame.

```python
df = pd.read_csv("winequality-red.csv")
```

```python
df.dtypes
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.count()
```

```python
print(df.isnull().sum())
```

## 3. Boxplot

A boxplot with one categorical(x-axis) and one continous variable (y-axis)

### Alcohol Vs Quality

```python
sns.boxplot(x='quality', y='alcohol', data = df)
```

We can remove the outliers by passing an argument, showoutliers=False

```python
sns.boxplot(x='quality', y='alcohol', data = df, showfliers=False)
```

The higher the alcohol concentration is, the higher the quality of the wine

## 4. Coorelation Matrix

A coorelation matrix of all the variables

```python
plt.figure(figsize=(20, 17))
matrix = np.triu(df.corr())
sns.heatmap(df.corr(), annot=True,
            linewidth=.8, mask=matrix, cmap="rocket");
```
